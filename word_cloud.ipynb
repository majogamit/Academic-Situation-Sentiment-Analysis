```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Read the CSV dataset
df = pd.read_csv('All_Data.csv')

# Extract hashtags from the 'text' column
hashtags = [word for sentence in df['text'] for word in sentence.split() if word.startswith('#')]

# Count the frequency of each hashtag
hashtag_counts = Counter(hashtags)

# Get the top 5 most frequently used hashtags
top_hashtags = hashtag_counts.most_common(10)

# Extract the hashtags and their counts into separate lists
hashtags = [hashtag[0] for hashtag in top_hashtags]
counts = [hashtag[1] for hashtag in top_hashtags]

for i in range(10):
    print(hashtags[i]+ ' - ' + str(counts[i]) + '\n')

# Plot a bar chart of the top 5 hashtags

plt.bar(hashtags, counts, color='pink')
plt.xlabel('Hashtags')
plt.ylabel('Frequency')
plt.plot(color='pink')
plt.title('Top 10 most frequently used hashtags')
plt.xticks(rotation=90)
plt.show()

data = {'Hashtag': hashtags[:10],
        'Count': counts[:10]}
df = pd.DataFrame(data)
df.head(10)
```

```python
#Importing Libraries

import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline

from wordcloud import WordCloud

#Importing Dataset

df = pd.read_csv("final_clean.csv")

#Checking the Data

df.head()

#Creating the text variable

text2 = " ".join(title for title in df.text)

# Creating word_cloud with text as argument in .generate() method

word_cloud2 = WordCloud(collocations = False, background_color = 'white').generate(text2)

# Display the generated Word Cloud
plt.figure(figsize = (10, 10), facecolor = None)
plt.imshow(word_cloud2, interpolation='bilinear')

plt.axis("off")

plt.show()
```

```python
academic_stopwords = ['join', 'like', 'f', 'us', 'classes', 'ease', 'pagodnakami', 'academic', 'academiceasenow', 'academicbreaknow', 'break', 'academicfreezenow', 'academicfreeze', 'freeze', 'ligtasnabalikeskwela', 'student', 'school', 'yung', 'di', 'co', 'naman', 'academicbreak', 'lang', 'po', 'nag', 'mag', 'pag', 'tapos', 'talaga', 'agad', 'week', 'nga', 'mo', 'muna', 'student', 'class', 'kayo', 'palang', 'nalang', 'students', 'call', 'need', 'year', 'eh', 'ta', 'u', 'yan', 'pala', 'niyo', 'parang', 'rin', 'din', 'si', 'acadbreak', 'wala', 'ba', 'day', 'kasi', 'sana', 'niyo', 'akong', 'nyo']
```

```python
tokenizer = ToktokTokenizer()
def remove_academic_stopwords(text):
    tokens = tokenizer.tokenize(text)
    tokens = [token.strip() for token in tokens]
    filtered_tokens = [token for token in tokens if token.lower() not in academic_stopwords]
    filtered_text = ' '.join(filtered_tokens)    
    return filtered_text

df['text']=df['text'].apply(remove_academic_stopwords)
```

```python
data = df['text'].tolist()

# Convert the strings to lowercase and split them into words
words = [word.lower() for sentence in data for word in sentence.split()]

# Count the frequency of each word
word_counts = Counter(words)

# Get the top 10 most frequently used words
top_words = word_counts.most_common(10)

# Extract the words and their counts into separate lists
words = [word[0] for word in top_words]
counts = [word[1] for word in top_words]

# Plot a bar chart of the top 10 words
plt.bar(words, counts)
plt.xlabel('Words')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.title('Top 10 most frequently used words')
plt.show()
```
