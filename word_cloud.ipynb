```python
from nltk.tokenize.toktok import ToktokTokenizer
```

```python
#Importing Libraries

import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline

from wordcloud import WordCloud

#Importing Dataset

df = pd.read_csv("final_clean.csv")

#Checking the Data

df.head()

#Creating the text variable

text2 = " ".join(title for title in df.text)

# Creating word_cloud with text as argument in .generate() method

word_cloud2 = WordCloud(collocations = False, background_color = 'white').generate(text2)

# Display the generated Word Cloud
plt.figure(figsize = (10, 10), facecolor = None)
plt.imshow(word_cloud2, interpolation='bilinear')

plt.axis("off")

plt.show()
```

```python
academic_stopwords = ['academic', 'academiceasenow', 'academicbreaknow', 'break', 'academicfreezenow', 'academicfreeze', 'freeze', 'ligtasnabalikeskwela', 'student', 'school', 'yung', 'di', 'co', 'naman', 'academicbreak', 'lang', 'po', 'nag', 'mag', 'pag', 'tapos', 'talaga', 'agad', 'week', 'nga', 'mo', 'muna', 'student', 'class', 'kayo', 'palang', 'nalang', 'students', 'call', 'need', 'year', 'eh', 'ta', 'u', 'yan', 'pala', 'niyo', 'parang', 'rin', 'din', 'si', 'acadbreak', 'wala', 'ba', 'day', 'kasi', 'sana', 'niyo', 'akong', 'nyo']
```

```python
tokenizer = ToktokTokenizer()
def remove_academic_stopwords(text):
    tokens = tokenizer.tokenize(text)
    tokens = [token.strip() for token in tokens]
    filtered_tokens = [token for token in tokens if token.lower() not in academic_stopwords]
    filtered_text = ' '.join(filtered_tokens)    
    return filtered_text

df['text_1']=df['text'].apply(remove_academic_stopwords)
```

```python
text2 = " ".join(title for title in df.text_1)

# Creating word_cloud with text as argument in .generate() method

word_cloud2 = WordCloud(collocations = False, background_color = 'white').generate(text2)

# Display the generated Word Cloud
plt.figure(figsize = (10, 10), facecolor = None)
plt.imshow(word_cloud2, interpolation='bilinear')

plt.axis("off")

plt.show()
```
